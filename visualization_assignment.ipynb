```python
import matplotlib.pyplot as plt
```

<h2> Primer Grafico de linea <h2>

```python
plt.plot([2000, 2005, 2010, 2015, 2020, 2023], [14.3, 14.5, 14.6, 14.8, 15.1, 15.3])
plt.title("Temperatura Global Promedio (°C)")
plt.xlabel("Año")
plt.ylabel("Temperatura (°C)")
plt.show()
```

Evolución de la temperatura global promedio (°C)

<h2> Segundo grafico de barras <h2>

```python
plt.bar(['Asia', 'África', 'Europa', 'América', 'Oceanía'], [4.7, 1.4, 0.7, 1.0, 0.04])
plt.title("Población por Continente (2024)")
plt.ylabel("Miles de millones")
plt.show()
```

<h4> Población de continentes (en miles de millones) <h4>

<h2> Tercer Grafico Histogram <h2>

```python
edades = [23, 45, 56, 67, 34, 45, 23, 36, 48, 52, 64, 29, 33, 37, 41, 22, 24, 25, 30]
plt.hist(edades, bins=5)
plt.title("Distribución de Edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

<h4>Distribución de edades en una población <h4>

<h2> Cuarto grafico Scatter plot <h2>

```python
pib = [1000, 5000, 10000, 20000, 40000, 60000]
esperanza_vida = [60, 65, 70, 75, 80, 83]

plt.scatter(pib, esperanza_vida)
plt.title("PIB per cápita vs Esperanza de Vida")
plt.xlabel("PIB per cápita (USD)")
plt.ylabel("Esperanza de Vida (años)")
plt.show()
```

<h4> Relación entre PIB per cápita y esperanza de vida <h4>

<h2> Quinto grafico Pie Chart <h2>

```python
fuentes = ['Petróleo', 'Carbón', 'Gas Natural', 'Renovables', 'Nuclear']
porcentajes = [31, 27, 23, 14, 5]

plt.pie(porcentajes, labels=fuentes, autopct='%1.1f%%')
plt.title("Distribución Mundial de Fuentes de Energía")
plt.show()
```

<h4>Uso de fuentes de energía en el mundo<h4>
